## 2. The question: decision, action, cost of a wrong call

**Decision**

Which content pages should be reviewed first for refresh?

**Who takes the action?**

A content editor or SEO specialist reviews the ranked list and decides which pages to update first.

**Cost of a wrong recommendation**

A false positive wastes editor time by recommending pages that do not need attention. A false negative misses pages that could lose traffic if they are not refreshed.

**Why ML?**

A simple rule cannot combine many interacting signals such as impressions, CTR, engagement, freshness, content age, and position. Machine learning can learn complex patterns and produce a better ranking while still supporting human decision-making.

## Quick Look at the Data

The starter dataset contains **30,000** content pages described by **44** features covering search performance, engagement, and content characteristics.

A preliminary inspection shows that **16,262 pages (54.21%)** are labeled as declining based on the `trend_direction` field, indicating that content decline is a common pattern in this dataset.

The average number of impressions over the last 90 days is **5,200.37**, while the average CTR is **0.511%**. These statistics suggest that pages vary considerably in visibility and performance, making it valuable to build a system that ranks which pages should be reviewed first for refresh instead of treating every page equally.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

df = pd.read_csv("/content/content_refresh_anonymized.csv")

print("Rows:", len(df))
print("Columns:", len(df.columns))

df["is_declining_label"] = df["trend_direction"] == "down"

print("\nDeclining pages:")
print(df["is_declining_label"].value_counts())

print("\nPercentage of declining pages:")
print((df["is_declining_label"].mean() * 100).round(2), "%")

print("\nAverage 90-day impressions:")
print(round(df["impressions_90d"].mean(), 2))

print("\nAverage CTR:")
print(round(df["ctr"].mean(), 3))
```
